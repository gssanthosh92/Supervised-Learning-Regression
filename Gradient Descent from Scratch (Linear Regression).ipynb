```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airline cost new.csv')
X = df['num_pass']
y = df['cost']
```

```python
plt.scatter(X,y)
plt.show()
```

```python
## build gradient descent algorithm

m=10
c=1400
L = 0.00001

epochs=1000
n = len(X)

for i in range(epochs):
    y_pred = m*X + c
    D_m = (-2/n)*sum(X*(y-y_pred)) ## slope of cost wrt m
    D_c = (-2/n)*sum(y-y_pred) ## slope of cost wrt c
    m = m-L*D_m ## update m based on slope
    c = c-L*D_c ## update c based on slope
    
    if (i%100) == 0:
        plt.scatter(X, y)
        plt.plot(X, y_pred)
        plt.show()
        
print ('m:  ', m, 'c:  ', c)
```
